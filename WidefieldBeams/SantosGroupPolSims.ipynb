```python
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u
from astropy import constants as c
import healpy as hp
```

```python
mask = hp.read_map('mask_tot_fskyQU90.fits')
```

```python
hp.mollview(mask)
plt.show()
```

```python
U = hp.read_map('pol_synch_50_200/map_U_nside512_noisecv_50_200_01_freq_60')
Q = hp.read_map('pol_synch_50_200/map_Q_nside512_noisecv_50_200_01_freq_60')
```

```python
hp.mollview(Q)
hp.mollview(U)
phs = np.arctan2(U,Q)
hp.mollview(np.arctan2(U,Q))
plt.show()
```

```python
def Cl(hpmap):
    return hp.alm2cl(hp.map2alm(hpmap))
cl_phs = Cl(phs)
cl_u = Cl(U)
cl_q = Cl(Q)
l = np.arange(len(cl_q))
```

```python
plt.semilogy(cl_phs)
plt.show()
```

```python
plt.figure()
plt.clf()
plt.loglog(l[1:],cl_u[1:])
plt.loglog(l[1:],cl_q[1:])
plt.xlim([5,1000])
plt.ylim([1e-10,1e-6])
plt.show()
```
